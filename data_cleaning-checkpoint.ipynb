```python
df = orders.merge(
    payments,
    on='order_id',
    how='left'
)
```

```python
df = df.merge(
    customers,
    on='customer_id',
    how='left'
)
```

```python
## Total Revenue

df['payment_value'].sum()
```

```python
## Revenue by State

state_revenue=(
    df.groupby('customer_state')['payment_value'].sum().sort_values(ascending = False)
)
state_revenue.head(10)
```

```python
state_revenue.head(10).plot(
    kind = 'bar',
    figsize = (10,5)
)
plt.title('Top 10 States by Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
## Top 10 Customers by Revenue

top_customers = (
    df.groupby('customer_unique_id')['payment_value'].sum().sort_values(ascending = False)
)
top_customers.head(10)
```

```python
## Visualization

top_customers.head(10).plot(
    kind = 'bar',
    figsize = (10,5)
)
plt.title('Top 10 Customers by Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
## Revenue Distribution

plt.figure(figsize = (8,5))
df['payment_value'].hist(
    bins = 50
)
plt.title('Revenue Distribution')
plt.xlabel('Payment Value')
plt.ylabel('Frequency')
plt.show()
```
